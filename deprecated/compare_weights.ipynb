```python
def count_unique_params_from_ordereddict(ordered_dict):
    seen_storages = set()
    total_params = 0
    total_zero_params = 0

    for name, values in ordered_dict.items():
        storage_id = id(values.untyped_storage())
        if storage_id not in seen_storages:
            seen_storages.add(storage_id)
            layer_params = values.numel()
            zero_params = (values == 0).sum().item()

            total_params += layer_params
            total_zero_params += zero_params

            print(f"{name} : {layer_params} parameters, "
                  f"{zero_params} are zero")
        else:
            print(f"Skipped {name} (duplicate storage)")

    zero_percentage = (total_zero_params / total_params) * 100 if total_params else 0
    print(f"\nUnique parameters in OrderedDict: {total_params}")
    print(f"Zero parameters in OrderedDict: {total_zero_params}")
    print(f"Percentage of zero (pruned) parameters: {zero_percentage:.2f}%")

    return total_params

unique_param_count = count_unique_params_from_ordereddict(model_2)
```
